### Best Performing Product by Profit Margin

```python
import pandas
```

#### Access the Data File

```python
url = 'https://raw.githubusercontent.com/Rizwan-Ahmed-Surhio/Accountants/main/5000%20Sales%20Records.csv'
df = pandas.read_csv(url)
```

```python

#define function and its names i.e. profit margin and input data

def profit_margin(df):
   
  # Create two empty list one that saves names of the product and other that saves profit margin numbers(%) after running the loop.
   names = []
   margins = []

   # extract products names from the database
   products = df['Item Type'].unique()
  
  #start a loop and pick a product from a list of products 
   for a in products:

        # search the product in the database 
        seleted_product= df.loc[df['Item Type'] == a]
        
        ###save product name in an empty list names using append method so we can use this list later on when creating a dataframe
        ### so we know margin relates to which product.
       
        names.append(a)
    
        ### calculate profit margin and similar to above save the margin number in margin list
        margins.append(round(seleted_product['Total Profit'].sum() / seleted_product['Total Revenue'].sum() * 100,3))
        
   # Now we are outside of loop
   ## Create an empty data frame name margin_df and created two columns based on two lists created above.
   margin_df = pandas.DataFrame()
   margin_df["Products"] =names
   margin_df["Margins"] =margins

   return (margin_df)   
        

 
```

### Use function profit_margin for desired database and save the results in df_products_margins


```python
df_products_margins = profit_margin(df)
```

```python
df_products_margins
```

### Sort from highest to lost profit margins to see highest at the top.

```python
df_products_margins.sort_values(by=["Margins"], ascending=False)
```

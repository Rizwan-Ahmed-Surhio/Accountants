```python
import pandas as pd
import pandas_datareader as wb
```

```python
## Setting the Dates
start = '2020-6-30'
end  = '2020-6-30'
```

```python
## Materialiy
materiality = abs(1000)
```

```python
## Load the CSV file and select symbol column for collecting data
url = 'https://raw.githubusercontent.com/Rizwan-Ahmed-Surhio/Accountants/main/Python%20for%20Accountants%20III/data/client_list.csv'
client_list = pd.read_csv(url)
```

```python
client_list
```

```python
stocks = (client_list.symbol)
```

```python
stocks
```

```python
# Example for one stock data . WOW.AX
woolworth = wb.DataReader('wow.ax', 'yahoo', start= start, end = end)
woolworth
```

```python
## Collect data from yahoo finance for all symbols and storing values in data
data = round((wb.DataReader(stocks, 'yahoo', start= start, end = end)['Close']).T,3)
```

```python
data
```

```python
data.drop(data.columns[0], axis = 'columns', inplace = True)
```

```python
data
```

```python
data.columns = ['AuditPrices']
```

```python
data
```

```python
client_list
```

```python
client_list.set_index('symbol', inplace=True)
```

```python
client_list
```

```python
client_list['AuditPrices'] = data['AuditPrices']
client_list
```

```python
client_list['PriceDifference'] = round(client_list['prices'] - client_list['AuditPrices'],2)
client_list
```

```python
client_list['AuditBalanceDifference'] = abs(round(client_list['PriceDifference'] * client_list['quantity'],2))
client_list
```

```python
client_list['Material?'] = ["" if x <=  materiality else 'Material' for x in client_list['AuditBalanceDifference']]
client_list
```

```python
client_list.to_csv('Investment_Testing.csv', index=False)
```
